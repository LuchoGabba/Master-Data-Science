## Clustering

### K-Means
Un modelo muy utilizado para clusterizaci贸n es el que vamos a ver a continuaci贸n k-Means (Ojo no confundir con KNN que es un m贸delo de clasificaci贸n supervisado).

La idea es agrupar los datos en funci贸n a la similitud que haya entre ellos en distintos clusters. Debemos definir la cantidad de clusters que queremos que el algoritmo encuentre (K).

Algoritmo:

* Asignar centroides (centro del cluster). En la primera iteraci贸n va a ser al azar luego se calcularan tomando el promedio de los puntos de cada cluster (figura b)
* Calcular qu茅 punto pertenece a cada cluster
* Promediar todos los puntos en cada dimensi贸n para recalcular el centroide
* Repetir hasta que los puntos dejen de cambiar de cluster, de tal manera que cada punto va a estar mas cerca del centroide de su cluster antes que cualquier otro centroide.

Este proceso utiliza el algor铆tmo 'Expectation-maximization (E-M)'. En este caso los pasos ser铆an:

* Expectation step: asignar cada punto al centroide m谩s cercano
* Maximization step: reasignar un centroide que sea el promedio de los nuevos puntos y evaluar si es mejor que el anterior

### DBSCAN

Otro algor铆tmo de clustering que resuelve algunas de las limitaciones que presenta K-means es DBSCAN. Un algoritmo que clusteriza por densidad de puntos en el espacio.

Para esto el algor铆tmo busca los casos centrales donde hay gran densidad y los utiliza para determinar el centro de cada cluster.

DBSCAN trabaja con distintas geometr铆as al contrario de K-means.

DBSCAN-density-data.svg.png

Los par谩metros que van a definir la densidad l铆mite para clusterizar son `min_samples` y `eps`.  A mayor min_samples o menor eps vamos a estar buscando una mayor densidad para generar los cluster.
* Min samples refiere a la cantidad m铆nima de casos agrupados por zona central que va a encontrar el algor铆tmo. Controla la tolerancia del algoritmo al ruido (cuanto mas ruidoso el dataset es mejor aumentar dicho par谩metro).
* Eps refiere a la distancia del resto de los casos a aquellos centrales. Es importante evaluar su valor ya que influye de gran manera en los resultados del algoritmo. Si el valor de eps es demasiado peque帽o la data puede ser clasificada como -1 caso contrario toda los datos pueden ser agrupados en un solo cluster erroneamente.

1. It starts with an arbitrary starting data point that has not been visited. The neighborhood of this point is extracted using the maximum distance eps.
2. If its neighborhood contains a sufficient number of points according to min samples, the clustering begins. That starting point becomes the core point of the cluster as well. Otherwise, the point is assumed as noise. Later on, this point can probably be in the neighborhood of another point and hence be a part of a cluster. In either case, this point is marked as a visited point.
3. The points in the neighborhood of the core point are then used to search their respective neighborhood points since these points are still new un-visited points.
4. This process of steps 2 and 3 is repeated until all points in the cluster have been visited and labeled forming the density-connected cluster.
5. Then, the new un-visited point in the data set is retrieved and the algorithm repeats through steps 1 to 4 until all points have been visited and become either noise or part of a cluster.

```python
import numpy as np

from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

generamos datos


```python
centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(
    n_samples=750, centers=centers, cluster_std=0.4, random_state=0
)

X = StandardScaler().fit_transform(X)
```

Utilizamos DBSCAN para encontrar los clusters

```python
db = DBSCAN(eps=0.3, min_samples=9).fit(X)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print('vemos algunas m茅tricas para determinar la performance de nuestro modelo')
print("Estimado n煤mero de cluster: %d" % n_clusters_)
print("Estimado n煤mero de puntos ruidosos: %d" % n_noise_)

# Homogeneity: Measures if each cluster contains only members of a single class. A value close to 1 indicates good homogeneity.
print("homogeneidad: %0.3f" % metrics.homogeneity_score(labels_true, labels))

#Completeness: Measures if all members of a given class are assigned to the same cluster. A value close to 1 indicates high completeness.
print("completitud: %0.3f" % metrics.completeness_score(labels_true, labels))

#V-measure: Harmonic mean of homogeneity and completeness. It's a compromise between the two.
print("V-measure: %0.3f" % metrics.v_measure_score(labels_true, labels))

print("Ajuste Rand Index: %0.3f" % metrics.adjusted_rand_score(labels_true, labels))
print("Ajuste Mutual Information: %0.3f" % metrics.adjusted_mutual_info_score(labels_true, labels))
print("Coeficiente Silhouette: %0.3f" % metrics.silhouette_score(X, labels))
```

```python
# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = labels == k

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=14,
    )

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=6,
    )

plt.title("Estimado n煤mero de clusters: %d" % n_clusters_)
plt.show()
```

- Como vemos este modelo de clusterizaci贸n encuentra por si solo la mejor cantidad de clusters. Sin embargo esto se puede ir modificando a trav茅s de los par谩metros que vimos. Pero no es necesario defenir el n煤mero de clusters como pasaba en K-means.
- DBSCAN es un modelo determin铆stico lo que indica que siempre va a llegar al mismo resultado al contrario de kmeans donde la aleatoreidad de la inicializaci贸n puedo llevarnos a diferentes tipos de resultados
- Una mejora de este algoritmo es HDBSCAN

### Evaluando calidad de clusterizaci贸n.

Evaluar la calidad de la clusterizaci贸n no es tan sencillo como cuando teniamos un target. En este caso entender si estamos haciendo las cosas bien sin tener una 'respuesta' a la cual queremos llegar va a depender de ciertas m茅tricas que nos indique que tan bien est谩 ajustando el algor铆tmo a los datos.
Podemos hablar por ejemplo de que tan bien se separan los clusters, o si los elementos de un cluster se parecen m谩s entre si que al resto utilizando alguna m茅trica de similaridad.
Mencionamos algunas m茅tricas:

- **Inertia**:
En el caso del m茅todo K-means la m茅trica utilizada para evaluar es **inertia**, que es la m茅trica que el mismo modelo trata de optimizar. A menor valor mejores m茅tricas.
- **Rand index**:
Si bien estos m茅todos son de aprendizaje no supervisado podemos tener el caso donde sabemos los grupos pertenecientes. En este caso podemos utilizar rand index que nos va a dar un score. Similar cuando trabajamos con clasificaci贸n
- **Coeficiente Silhouette**:
este coeficiente describe que tan bien agrupan los clusters. A mayor coeficiente vamos a encontrar grupos mejor definidos. Esta m茅trica trabaja con la distancia entre un punto y el resto de los puntos de su propio grupo y la distancia entre un punto y los puntos de grupo m谩s cercano que no es el de s铆 mismo. Este valor esta normalizado entre -1 y 1. Este coeficiente funciona mejor con formas convexas.
- **Calinski-Harabasz Index**: A mayor valor mejores grupos esta encontrando nuestro m茅todo de clusterizaci贸n. Esta m茅trica trabaja con la disperci贸n de los clusters.

### Principal Componet Analysis: PCA

Este metodo no supervisado para la reducci贸n de dimensionalidad es muy flexible y da buenos resultados para la mayor铆a de los tratamientos. Vamos a trabajar con la introducci贸n brindada por la librer铆a [Scikit-Learn](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html)



```python
#generamos datos sint茅ticos: (de mentira :p)
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal')
```

Vemos una relaci贸n lineal entre los datos.

Para realizar una reducci贸n de dimensionalidad este m茅todo usa la matriz de covarianza entre las variables y la descompone en una matr铆z autovectores para entender como explicar la relaci贸n entre los puntos con los vectores fundamentales.

La matr铆z de covarianza es la matriz donde los elementos i,j de la matriz representan la covarianza entre la variable i y la variabe j.


Algunas propiedades para tener encuenta sobre la covarianza:

* La covarianza de un vector contra si mismo es igual a su varianza.
* Cada elemento de la diagonal principal representa la varianza de la variable correspondiente
* La matriz de covarianza es sim茅trica


- PCA:
Evalua la relaci贸n entre variables utilizando la matr铆z de covarianza y entiende los principales componentes que describe nuestros datos a trav茅s de sus autovalores. El peso que tiene cada autovector para describir la naturaleza de los datos es dado por su autovalor asociado.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
print(pca.components_)
```

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0, color='black')
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal')
```

#### Implementacion: de R2 a R

* Obtenemos matriz de covarianza. Ejemplo R3:

  los valores positivos van a indicar relaciones positivas (crecen o decrecen de la misma manera) y viceversa.

* Obtenemos autovectores y autovalores para entender los componentes principales y su importancia.

  Si recordamos un poco de lgebra los autovectores son los vectores principales que explican un espacio. De otra manera, son los m铆nimos vectores necesarios que al combinarlos linealmente pueden recrear todo el subespacio. En nuetro caso todo el dataset. Estos vectores no estan correlacionados entre s铆. Aquellos autovectores con autovalores m谩s grandes son los m谩s determinantes del subespacio. Los que poseen mayor informaci贸n o **explican la mayor cantidad de varianza**. Este 煤ltimo concepto es importante: Relacionamos informaci贸n con varianza ya que a mayor dispersi贸n de datos estamos teniendo mayor cantidad de informaci贸n. Si los datos estan muy compactos en pocos valores, o sea poseen poca dispersi贸n, estamos teniendo poca informaci贸n; pocos valores.

La idea detras de PCA es eliminar esos componentes que no otorgan informaci贸n significativa sobre nuestros datos. En el caso visto m谩s arriba vemos que el autovector de menor magnitud es aquel que es eliminado para hacer la reducci贸n de dimensionalidad y nos quedamos con el mayor para recrear nuestro subespacio con una dimensi贸n menor.

```python
# Normalizamos solo extrayendo la media por fines pr谩cticos (va a estar centrado en 0)
X_meaned = X - np.mean(X , axis = 0)

# Matriz de covarianza
cov_mat = np.cov(X_meaned , rowvar = False)

# Autovalores y autovectores de dicha matriz
autovalores, autovectores = np.linalg.eigh(cov_mat)

# Ordenamos los autovalores de manera descendiente
sorted_index = np.argsort(autovalores)[::-1]
sorted_autovalores = autovalores[sorted_index]

# Ordenamos los autovectores en relaci贸n a los autovalores
sorted_autovalores = autovectores[:,sorted_index]

# Elegimos los primeros n valores donde n es la dimensi贸n de nuestro nuevo subespacio

n_dim = 1
autovalores_subset = sorted_autovalores[:,0:n_dim]

# por 煤ltimo hacemos una transformaci贸n lineal de los datos que tenemos utilizando el producto punto
X_redux = np.dot(autovalores_subset.transpose(),X_meaned.transpose()).transpose()
```

### Reducci贸n de dimensionalidad: TSNE y UMAP

Para salir del espacio euclideano podemos utilizar estos dos extendidos m茅todos para reducir dimensionalidad.

t-SNE ([t-distributed Stochastic Neighbor Embedding](https://www.jmlr.org/papers/volume9/vandermaaten08a/vandermaaten08a.pdf))


Este m茅todo llega para resolver el problema que tiene PCA para trabajar con espacios no euclideanos que no podemos abarcar con lgebra lineal.
Por ende es m谩s complejo que PCA y vamos a tocar los conceptos principales unicamente.

t-SNE trabaja en primer lugar calculando la distribuci贸n de probabilidad que representa similaridad entre puntos vecinos. Una distribuci贸n que responde a que tan probable es que elija este otro punto como vecino. Esta distribuci贸n el modelo la supone Gauseana.

**C贸mo realiza entonces la reducci贸n de dimensionalidad**

Al encontrar distribuciones parecidas entre distintos grupos de vecinos (s铆, es una clusterizaci贸n) el modelo va a tratar de unirlos en una nueva distribuci贸n de probabilidad que le permita deshacerse de una de las dimensiones que describen a los datos. La nueva distribuci贸n que utiliza este m茅todo es *T de student*. De aqu铆 el nombre.

Este proceso se optimiza iterativamente para encontrar los mejores grupos y las mejores distribuciones que los expliquen. Para esto se utiliza una distancia estad铆stica muy extendida como m茅trica en machine learning conocida como [divergencia Kullback-Leibler](https://en.wikipedia.org/wiki/Kullback%E2%80%93Leibler_divergence)

```python
from numpy.random import RandomState
import matplotlib.pyplot as plt
from matplotlib import ticker

# unused but required import for doing 3d projections with matplotlib < 3.2
# import mpl_toolkits.mplot3d  # noqa: F401

from sklearn import manifold, datasets

rng = RandomState(0)

n_samples = 1500
S_points, S_color = datasets.make_s_curve(n_samples, random_state=rng)
```

```python
#recreamos un espacio no euclideano
def plot_3d(points, points_color, title):
    x, y, z = points.T

    fig, ax = plt.subplots(
        figsize=(6, 6),
        facecolor="white",
        tight_layout=True,
        subplot_kw={"projection": "3d"},
    )
    fig.suptitle(title, size=16)
    col = ax.scatter(x, y, z, c=points_color, s=50, alpha=0.8)
    ax.view_init(azim=-60, elev=9)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.zaxis.set_major_locator(ticker.MultipleLocator(1))

    fig.colorbar(col, ax=ax, orientation="horizontal", shrink=0.6, aspect=60, pad=0.01)
    plt.show()


def plot_2d(points, points_color, title):
    fig, ax = plt.subplots(figsize=(3, 3), facecolor="white", constrained_layout=True)
    fig.suptitle(title, size=16)
    add_2d_scatter(ax, points, points_color)
    plt.show()


def add_2d_scatter(ax, points, points_color, title=None):
    x, y = points.T
    ax.scatter(x, y, c=points_color, s=50, alpha=0.8)
    ax.set_title(title)
    ax.xaxis.set_major_formatter(ticker.NullFormatter())
    ax.yaxis.set_major_formatter(ticker.NullFormatter())


plot_3d(S_points, S_color, "Original S-curve samples")
```

```python
t_sne = manifold.TSNE(
    n_components=2,
    learning_rate="auto",
    perplexity=30,
    n_iter=250,
    init="random",
    random_state=rng,
)
S_t_sne = t_sne.fit_transform(S_points)

plot_2d(S_t_sne, S_color, "T-distributed Stochastic  \n Neighbor Embedding")
```
