Take this data as inputs

```python
# Libraries
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Random Data
X = np.linspace(-5 , 5, 1000).reshape(1000, 1)
y_clean = np.sin(3 * X) + X

# Add some noise to generate a non perfect function
y_noise = np.random.normal(size=len(X)).reshape(y_clean.shape)
y = y_clean + y_noise
```

Perform multiple Decission Tree Regressors in each model use the following values for the parameter max_depth:
* 3
* 4
* 5
* 7
* 30
* 50

Check for training and test when the model starts to produce overfitting

## SOLUTION

```python
plt.scatter(X, y)
plt.xlabel("Set X")
plt.ylabel("Target Y")
plt.show()
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5)
```

https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html

## MAX_DEPTH = 3

```python
# Training a Decission Tree model
regr = DecisionTreeRegressor(max_depth=3)
regr.fit(X_train, y_train)

y_train_pred = regr.predict(X_train)
y_test_pred = regr.predict(X_test)

print('R^2 in train : {}'.format(np.round(r2_score(y_train, y_train_pred),2)))
print('R^2 in test: {}'.format(np.round(r2_score(y_test, y_test_pred),2)))

X_fit = np.linspace(-5, 5, 1000).reshape(1000, 1)
y_fit = regr.predict(X_fit)

plt.scatter(X_test, y_test, label="Original Data", alpha=0.4, s=10)
plt.plot(X_fit, y_fit, c="red", label="Predictions")
plt.legend()
plt.xlabel("Set X")
plt.ylabel("Target Y")
plt.show()
```
