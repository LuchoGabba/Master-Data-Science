# REGRESSION KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import GridSearchCV

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Random Data
np.random.seed(0)
X = np.sort(20 * np.random.rand(200,1), axis=0)
y = np.sin(X)[:, 0]
y_noise = np.sin(X)[:, 0] + np.random.normal(size=X.shape[0],scale=0.20)
```

```python
plt.xlabel("X set")
plt.ylabel("Target")
plt.scatter(X, y.ravel())
plt.show()
```

https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html

```python
clf = KNeighborsRegressor()
```

```python
clf
```

```python
clf.fit(X, y_noise)
```

```python
# Generating new data for predictions
X_axis = np.linspace(X.min(),X.max(), 200).reshape(200, 1)
```

```python
y_axis_pred = clf.predict(X_axis)
```

```python
plt.xlabel("X")
plt.ylabel("Target")
plt.scatter(X, y_noise.ravel())
plt.plot(X_axis, y_axis_pred, c="red")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print('MSE: %.2f' % mean_squared_error(y, y_axis_pred))
print('R_2: %.2f' % r2_score(y, y_axis_pred))
```
