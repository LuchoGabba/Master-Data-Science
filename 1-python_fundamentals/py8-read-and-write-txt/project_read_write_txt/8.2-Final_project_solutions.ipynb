## Read and clean function

Let us introduce the cleaning characters as arguments of the function

```python
def read_and_clean(file_name, comment_character = '', *comments):
    '''
    Open a file by calling it by its name as the function's argument. 
    The name must be a string that includes the directoy, if any.
    In addition, the function takes as many characters as comments are present with that format in the file,
    and thus detects them and skips them when reading.
    The output is a list of all valid entries from the original .txt file.
    '''
    with open(file_name, "r", encoding='utf-8') as data_file:

# Non-blank lines. Stores not blank entries from data_file in a list
        lines = list(line for line in (l.strip() for l in data_file) if line)
        
# Clean lines with comments
        list_char = [comment_character]
        for arg in comments: list_char.append(arg) 
        
        valid_entries = []
        for item in lines:
            if item.startswith(tuple(list_char)):
                pass
            else:
                valid_entries.append(item)
                
# print a sum up of the returned object              
        print(f'The function returns a list with {len(valid_entries)} valid entries.')
    return valid_entries
```

### Features selector function

There are these names to srip:
- \~\~Todoyu~~
- burrrd.
- xmarks/
- Wickr...

Some URLs end with " / "

```python
def features_picker_and_cleaner(list_of_lines, feature, leftover_character = '', *leftover_characters):

    for arg in leftover_characters:
        leftover_character += arg
    
    list_of_one_features = list(map(lambda iterable : iterable.split(',')[feature].strip(leftover_character), list_of_lines))
    
    return list_of_one_features
```

### Occurrence counter dictionary creation function

```python
def values_and_frequencies(list_of_lines, min_num_repetitions):
    '''
    hi
    '''
    duplicates = {}
    for i in list_of_lines:
        if list_of_lines.count(i) >= min_num_repetitions:
            duplicates[i] = list_of_lines.count(i)

    #for k,v in duplicates.items():
    #    print('\t',k, v)  

    print('- Number of entrances (duplicates if second parameter is 2 or more): ', len(duplicates))
    return duplicates
```

## Function to filter names

```python
def filter_names(list_of_lines):
    '''
    Takes the list with all the entries and 
    generate a .txt file with half of the names 
    sorted alphabetically
    '''
    names_list = features_picker_and_cleaner(list_of_lines, 0, ' ', '~', '.', '/')
    names_list.sort()

# Preserve unique names 
    names_set = set(names_list)
    print(f'From the initial list with {len(names_list)} elements, {len(names_set)} are unique.', )
    
#    for i in names_set:
#        print(i)
    
    names_list = names_list[int(len(names_list)//2):]
    
    print(f'So the produced file has {len(names_list)} entries, which are\n\n', names_list)
    with open('Files/data_names.txt', 'w', encoding='utf-8') as data_file:    
        data_file.write('\n'.join(names_list))
```

## Function to extract the webpage domain

```python
def domain_extraction(list_of_lines):
    url_list = features_picker_and_cleaner(list_of_lines, 1, ' ', '~', '.', '/')
    print('url_list_size: ', len(url_list))#, url_list)

# Unique URLs:
    url_set = set(url_list)
    print('url_set_size: ', len(url_set))

# Repeated URLs:
    url_duplicates = values_and_frequencies(url_list, 2)
    
# Hay entradas que empiezan con http:// y otras con https://
# url_set counts both as different domains
    domains_list = list()
    for element in url_set:
        element = element.split('//', 1)[1]
        if element.startswith('www.'):
            domains_list.append(element.lstrip('www.'))
        else:
            domains_list.append(element)
    
    print('domains_list: ',len(domains_list))

# Unique domains:
    domains_set = set(domains_list)
    print('domains_set: ',len(domains_set))

# Repeated domains:
    domains_duplicates = values_and_frequencies(domains_list, 2)

    #print(len(domains_set),len(filtered_domains_list),domains_set)
    return (list(domains_set))
```

## Function to extract shared characteristics in the description

```python
def characteristics_organization(list_of_lines):

    characteristics_list = features_picker_and_cleaner(list_of_lines, 2, ' ', '~', '-')
    print(len(characteristics_list))#,characteristics_list)
    
    characteristics_dictionary = values_and_frequencies(characteristics_list, 1)
    
    characteristics_dictionary['n/a'] = characteristics_dictionary.pop('')
    
    return characteristics_dictionary
```
